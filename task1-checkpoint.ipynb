```python
%matplotlib inline

from pandas import read_csv, DataFrame, Series
import matplotlib.pyplot as plt
import pandas as pd

data = read_csv('pokemon.csv')
useful_cols = ['lvl_100_exp','gender_rate']
data[useful_cols].head()


```

Посмотрим как распределен  экспириенс 100 лвла. 

```python
df = data[['lvl_100_exp'] + ['id']]
df.groupby('id').sum().plot(kind='bar')


plt.figure(figsize=(5, 5))
plt.plot( data['id'][:649],data['lvl_100_exp'][:649], 'bo');
```

Сильно обобщив : вырисовывается 3 группы покемонов ( низковатый опыт,средний и высокий - к березкам тянутся)



Рассмотрим колонку про гендерный уровень( пусть "gender_rate" это про то,что покемоны мультиполвые).

```python
counts_ = data['gender_rate'].value_counts()
counts_.plot(kind='bar')
```

Очень интересно, переобладающее количество покемонов с гендером "4".Будет ли влиять отнесенность к этому гендеру на опытность покемона?

```python
data2= data[data.gender_rate == 4]
strange2 = data2[['lvl_100_exp'] + ['id']]

strange2.groupby('id').sum().plot(kind='bar')



```

Нет,вроде есть такие низкие и высокие показатели.

Посмотрим теперь на каждый гендер в отдельности и его  опыт 100ого уровня.

```python
strange2 = data[['lvl_100_exp'] + ['gender_rate']]

strange2.groupby('gender_rate').plot(kind='bar')
```
